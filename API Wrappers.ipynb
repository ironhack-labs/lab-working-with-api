```python
artists = ['Pearl Jam','John Butler Trio', 'Public Enemy']
```

```python
def artist_dict(artists):
    my_dict = {}
    for artist in artists:
        key = artist
        values = sp.search(q=artist, limit=5)
        my_dict[key] = values
    return my_dict
```

```python
my_artist_dict = artist_dict(artists)  # 'artists' is your list of artist names
```
