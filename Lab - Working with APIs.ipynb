```python
def flatten_dict(dd, separator='_', prefix=''):
        return {f"{prefix}{separator}{k}" if prefix else k: v
                for kk, vv in dd.items()
                for k, v in (flatten_dict(vv, separator, kk).items() if isinstance(vv, dict) else {kk: vv}.items())}
```
