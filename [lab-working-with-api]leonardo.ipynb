```python
def flight_info(flight_data):
    """
    Analyzes flight data to identify flights.
    
    Parameters:
    - flight_data: A dictionary containing information about flight quotes, places, and carriers.
    
    Returns:
    - A pandas DataFrame containing flights infos.
    """
    
    # Initialize an empty list to collect flight information
    all_flight_informations = []
        
    # Create mappings from PlaceId and CarrierId to their respective names for easy lookup
    place_mapping = {place['PlaceId']: place['Name'] for place in flight_data['Places']}
    carrier_mapping = {carrier['CarrierId']: carrier['Name'] for carrier in flight_data['Carriers']}
    
    # Iterate over each flight quote to extract and transform relevant information
    for quote in flight_data['Quotes']:
        # Extracting flight details from the quote
        price = quote['MinPrice']
        origin_id = quote['OutboundLeg']['OriginId']
        carrier_id = quote['OutboundLeg']['CarrierIds'][0]
        destination_id = quote['OutboundLeg']['DestinationId']
        departure_date = quote['OutboundLeg']['DepartureDate']

        # Using mappings to get human-readable names for origin and destination
        origin_name = place_mapping[origin_id]
        destination_name = place_mapping[destination_id]
        carrier_name = carrier_mapping[carrier_id]
        
        # Compiling extracted information into a dictionary
        flight_information = {
            'Flight Id': quote['QuoteId'],
            'Price': price,
            'Origin': origin_name,
            'Destination': destination_name,
            'Carrier': carrier_name,
            'Departure Date': departure_date
        }
        
        # Adding the compiled flight information to the list
        all_flight_informations.append(flight_information)
    
    # Converting the list of flight information into a pandas DataFrame
    df = pd.DataFrame(all_flight_informations)
    
    # Returning the DataFrame
    return df
```

```python
def flight_info_same_price(flight_data):
    """
    Analyzes flight data to identify flights with the same price but different departure dates.
    
    Parameters:
    - flight_data: A dictionary containing information about flight quotes, places, and carriers.
    
    Returns:
    - A pandas DataFrame containing flights that have the same price but depart on different dates.
    """
    
    # Initialize an empty list to collect flight information
    all_flight_informations = []
        
    # Create mappings from PlaceId and CarrierId to their respective names for easy lookup
    place_mapping = {place['PlaceId']: place['Name'] for place in flight_data['Places']}
    carrier_mapping = {carrier['CarrierId']: carrier['Name'] for carrier in flight_data['Carriers']}
    
    # Iterate over each flight quote to extract and transform relevant information
    for quote in flight_data['Quotes']:
        # Extracting flight details from the quote
        price = quote['MinPrice']
        origin_id = quote['OutboundLeg']['OriginId']
        carrier_id = quote['OutboundLeg']['CarrierIds'][0]
        destination_id = quote['OutboundLeg']['DestinationId']
        departure_date = quote['OutboundLeg']['DepartureDate']

        # Using mappings to get human-readable names for origin and destination
        origin_name = place_mapping[origin_id]
        destination_name = place_mapping[destination_id]
        carrier_name = carrier_mapping[carrier_id]
        
        # Compiling extracted information into a dictionary
        flight_information = {
            'Flight Id': quote['QuoteId'],
            'Price': price,
            'Origin': origin_name,
            'Destination': destination_name,
            'Carrier': carrier_name,
            'Departure Date': departure_date
        }
        
        # Adding the compiled flight information to the list
        all_flight_informations.append(flight_information)
    
    # Converting the list of flight information into a pandas DataFrame
    df = pd.DataFrame(all_flight_informations)
    
    # Converting 'Departure Date' from string to datetime.date for easier comparison
    df['Departure Date'] = pd.to_datetime(df['Departure Date']).dt.date

    # Filtering the DataFrame for flights with the same price but different departure dates
    result = df.groupby('Price').filter(lambda x: len(x['Departure Date'].unique()) > 1)
    
    # Returning the filtered DataFrame
    return result
```
