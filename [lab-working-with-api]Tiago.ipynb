```python
from datetime import datetime

def find_flights_same_price(start_date, end_date):
    
    Places = pd.json_normalize(flights['Places'])
    Carriers = pd.json_normalize(flights['Carriers'])
    Quotes = pd.json_normalize(flights['Quotes'])
    Quotes['OutboundLeg.CarrierIds'] = Quotes['OutboundLeg.CarrierIds'].apply(lambda x: int(x[0]))
    
    quotes_and_places_dest=pd.merge(Quotes, Places, left_on='OutboundLeg.DestinationId', right_on ='PlaceId')
    quotes_and_places_dest = quotes_and_places_dest.rename(columns={"Name": "Destination_Name", "SkyscannerCode": "Destination_Code", "CityName":"Destination_City_Name", "CityId":"Destination_City_id",
                                                                              "CountryName":"Destination_Country_Name","OutboundLeg.DepartureDate":"DepartureDate","OutboundLeg.CarrierIds": "CarrierIds",
                                                                              "OutboundLeg.OriginId":"OriginId","OutboundLeg.DestinationId":"DestinationId"})
    quotes_and_places = pd.merge(quotes_and_places_dest, Places, left_on='OriginId', right_on ='PlaceId')
    quotes_and_places=quotes_and_places.drop(columns=['PlaceId_y','Type_x','IataCode_x','IataCode_y','Type_y'])
    quotes_and_places= quotes_and_places.rename(columns={"Name": "Origin_Name", "SkyscannerCode": "Origin_Code", "CityName":"Origin_City_Name", "CityId":"Origin_City_id",
                                                                                  "CountryName":"Origin_Country_Name"})
    quotes_and_places_and_carriers=pd.merge(quotes_and_places, Carriers, left_on='CarrierIds', right_on ='CarrierId')
    quotes_and_places_and_carriers=quotes_and_places_and_carriers.rename(columns={"Name": "Airline"}) 
    quotes_and_places_and_carriers=quotes_and_places_and_carriers.drop(columns=['PlaceId_x','CarrierIds'])
    quotes_and_places_and_carriers['DepartureDate'] = pd.to_datetime(quotes_and_places_and_carriers['DepartureDate'])
    
    quotes_and_places_and_carriers.set_index(quotes_and_places_and_carriers['DepartureDate'], drop=False, append=False, inplace=True)
    
    
    return pd.DataFrame({'Price USD':quotes_and_places_and_carriers['MinPrice'],
                        'Origin':quotes_and_places_and_carriers['Origin_Name'],
                        'Origin_Code':quotes_and_places_and_carriers['Origin_Code'],
                        'Origin_City_Name':quotes_and_places_and_carriers['Origin_City_Name'],
                        'Arrival':quotes_and_places_and_carriers['Destination_Name'],
                        'Arrival_Code':quotes_and_places_and_carriers['Destination_Code'],
                        'Arrival_City_Name':quotes_and_places_and_carriers['Destination_City_Name'],
                        'Company':quotes_and_places_and_carriers['Airline']})
```
